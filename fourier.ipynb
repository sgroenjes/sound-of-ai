```python
import librosa
import librosa.display
import scipy as sp
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
woah_file = "woah-mono.wav"
tool_file = "tool-mono.wav"
```

```python
woah, sr = librosa.load(woah_file)
tool, _ = librosa.load(tool_file)
```

```python
FRAME_LENGTH = 1024
HOP_LENGTH = 512

ft_woah = sp.fft.fft(woah)
mag_woah = np.absolute(ft_woah)
freq_woah = np.linspace(0, sr, len(mag_woah))

ft_tool = sp.fft.fft(tool)
mag_tool = np.absolute(ft_tool)
freq_tool = np.linspace(0, sr, len(mag_tool))
```

```python
plt.figure(figsize=(15,17))

plt.subplot(3,1,1)
plt.plot(freq_woah[:5000], mag_woah[:5000])
plt.title("Woah FT")
plt.xlabel("Freqency (Hz)")

plt.subplot(3,1,2)
plt.plot(freq_tool[:5000], mag_tool[:5000])
plt.title("Tool FT")
plt.xlabel("Freqency (Hz)")

plt.show()
```
